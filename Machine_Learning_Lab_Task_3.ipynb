```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras import layers
```

```python
data = pd.read_csv(r'/content/Nairobi Office Price Ex.csv')
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
target =['PRICE']
y = data[target]
y = y.to_numpy()
y
```

```python
features = data.drop('PRICE',axis = 1,)
features
```

```python
all_features = list(features.columns)
unwanted = [2,8]
for ele in sorted(unwanted, reverse = True):
    del all_features[ele]
all_features
```

```python
#Encode the data

ordinal = data['AMB']
ordinal_data = pd.DataFrame(ordinal.astype('category').cat.codes)


nominal = data[all_features]
nominal_data = pd.get_dummies(nominal)
ordinal_data
```

```python
#Concantenate features
encoded_data = pd.concat([ordinal_data,nominal_data,data['SIZE']],axis =1)
#Convert feature to numpy array
x = encoded_data.to_numpy()
x[0]
```

```python
# Define Sequential model with 3 layers
model = keras.Sequential(
    [
        layers.Dense(3, activation="tanh", name="layer1"),
        layers.Dense(4, activation="tanh", name="layer2"),
        layers.Dense(1, name="layer3"),
    ]
)
```

```python
model.compile(optimizer='adam', loss = 'mean_squared_error')
```

```python
model.fit(x,y, epochs = 10, batch_size = 2)
```
